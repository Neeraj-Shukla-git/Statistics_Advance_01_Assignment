Question 01. What is a random variable in probability theory?
- A random variable is a numerical outcome of a random process or experiment. It assigns a real number to each possible outcome in a sample space.

Question 02. What are the types of random variables?
- There are two main types:
 - Discrete Random Variable - Takes countable values (e.g., number of heads in coin tosses).
 - Continuous Random Variable - Takes any value within a given range (e.g., height, temperature).

Question 03. What is the difference between discrete and continuous distributions?
- Discrete distributions involve variables with distinct, separate values (e.g., 0, 1, 2...).
- Continuous distributions involve variables with infinitely many values in an interval (e.g., all values from 0 to 1).

Question 04. What are probability distribution functions (PDF)?
- A Probability Distribution Function describes the likelihood of each possible value of a random variable:
 - For discrete variables, it's called a probability mass function (PMF).
 - For continuous variables, it refers to the probability density function (PDF).

Question 05. How do cumulative distribution functions (CDF) differ from PDF?
- PDF gives the probability of a value at a specific point (for continuous, this is density).
- CDF gives the probability that a variable takes a value less than or equal to a specific value:
 - F(x)=P(X≤x)

Question 06. What is a discrete uniform distribution?
- A discrete uniform distribution assigns equal probability to all outcomes in a finite set.
- Example: Rolling a fair six-sided die.

Question 07. What are the key properties of a Bernoulli distribution?
- Only two outcomes: success (1) and failure (0).
- Single trial with probability p for success and 1-p for failure.
- Expected value: E[X]=p
- Variance: Var(X)=p(1-p)

Question 08. What is the binomial distribution, and how is it used in probability?
- A binomial distribution describes the number of successes in n independent Bernoulli trials with success probability p.

Question 09. What is the Poisson distribution and where is it applied?
- The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a known average rate (λ) and independent events.
- Used in: queue theory, network traffic, etc.

Question 10. What is a continuous uniform distribution?
- A continuous uniform distribution has constant probability density over an interval [a, b].
- PDF:f(x)= 1/ b - a, a≤x≤b

Question 11. What are the characteristics of a normal distribution?
- Bell-shaped and symmetric around the mean.
- Mean = Median = Mode.
- Defined by two parameters: mean (μ) and standard deviation (σ).
- 68-95-99.7 rule: about 68% of data within 1σ, 95% within 2σ, 99.7% within 3σ.

Question 12. What is the standard normal distribution, and why is it important?
- A standard normal distribution is a normal distribution with:
- Mean = 0
- Standard deviation = 1
 - It's important because many statistical techniques (like Z-tests) assume data is standardized.

Question 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- CLT states: Regardless of the population distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size increases (n > 30 typically).
- This allows for use of normal-based inference methods on various data types.

Question 14. How does the Central Limit Theorem relate to the normal distribution?
- It explains why the normal distribution arises so frequently in practice. The mean of many samples from any distribution tends to be normally distributed.

Question 15. What is the application of Z statistics in hypothesis testing?
- Z-statistics is used to test hypotheses about population means when population variance is known.
- It helps determine if observed data significantly deviates from the null hypothesis.

Question 16. How do you calculate a Z-score, and what does it represent?
- Z = X - μ / σ
- Where:
  - X = observed value
  - μ = mean
  - σ = standard deviation
  - Z-score represents how many standard deviations an element is from the mean.

Question 17. What are point estimates and interval estimates in statistics?
- Point estimate: A single value estimate of a parameter (e.g., sample mean).
- Interval estimate: A range of values (e.g., confidence interval) likely to contain the parameter.

Question 18. What is the significance of confidence intervals in statistical analysis?
- They provide a range of values that likely contain the true population parameter, offering a measure of uncertainty and reliability for estimates.

Question 20. How are Z-scores used to compare different distributions?
- Z-scores standardize values from different distributions, allowing direct comparison by expressing all values in terms of standard deviations from their means.

Question 21. What are the assumptions for applying the Central Limit Theorem?
- Independent and identically distributed (i.i.d.) random variables.
- Finite mean and variance.
- Large sample size (n ≥ 30 typically).

Question 22. What is the concept of expected value in a probability distribution?
- The expected value is the average value of a random variable over many trials.

Question 23. How does a probability distribution relate to the expected outcome of a random variable?
- The probability distribution describes how likely each outcome is, and the expected value is the weighted average of those outcomes based on their probabilities.
